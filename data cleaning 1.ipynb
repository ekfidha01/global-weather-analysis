## Data Cleaning

### cleaning numerical columns with charecters

```python
df.head(10)
```

```python
df.drop(['sunrise','sunset','moonrise', 'moonset','moon_phase','moon_illumination'], axis=1, inplace=True)
```

```python
df
```

```python
df.rename(columns={
    'air_quality_Carbon_Monoxide': 'co',
    'air_quality_Ozone': 'ozone',
    'air_quality_PM2.5': 'pm25',
    'air_quality_PM10': 'pm10'
}, inplace=True)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df['last_updated'] = pd.to_datetime(df['last_updated'])
```

```python
df['year'] = df['last_updated'].dt.year
df['month'] = df['last_updated'].dt.month
df['hour'] = df['last_updated'].dt.hour
df
```

```python
df = df[(df['temperature_celsius'] > -50) & (df['temperature_celsius'] < 60)]
df
```

```python
df['humidity'] = df['humidity'].astype(int)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df = df[(df['humidity'] >= 0) & (df['humidity'] <= 100)]
```

```python
df.columns = df.columns.str.strip() 
```

```python
df['last_updated'] = pd.to_datetime(df['last_updated'])
```

```python
df
```

```python
df['temp_category'] = pd.cut(df['temperature_celsius'],
                            bins=[-50, 15, 30, 60],
                            labels=['cold', 'moderate', 'hot'])
```
