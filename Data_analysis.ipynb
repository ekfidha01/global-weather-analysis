## Distribution of Temperature Globally

- Shows the most common temperature range and the overall spread of temperature values globally.

```python
plt.hist(df['celsius'], bins=20)
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Frequency')
plt.title('Global Temperature Distribution')
plt.show()
```

- X-axis → Temperature (Celsius)
- Y-axis → Frequency

- The histogram shows how temperature values are distributed across the dataset.
- Taller bars represent temperature ranges that occur more frequently.
- It helps identify the most common temperature range globally.
- It also shows variation and spread of temperature values.

## Global Temperature Distribution Year-wise

- Shows how temperature distribution varies across different years, including spread and extreme values.

```python
yearly = df.groupby('year')['celsius'].mean()

# Plot
plt.figure(figsize=(8,5))
plt.plot(yearly.index, yearly.values, marker='o')

plt.xlabel('Year')
plt.ylabel('Average Temperature (Celsius)')
plt.title('Global Temperature Trend Year-wise')

plt.show()
```

- X-axis → Year
- Y-axis → Avarage Temperature (Celsius)

- The graph shows the trend of temperature over the years.
- Each point represents the average temperature of a year.
- An upward trend indicates increasing temperature.
- It helps understand long-term climate patterns.

## Air Pollution Comparison

- Average Air Pollution Levels (PM2.5 vs PM10)

```python
pm25_avg = df['pm2.5'].mean()
pm10_avg = df['pm10'].mean()

# Data
pollutants = ['PM2.5', 'PM10']
values = [pm25_avg, pm10_avg]

# Plot
plt.bar(pollutants, values)

plt.xlabel('Pollutants')
plt.ylabel('Average Concentration')
plt.title('Global Air Pollution Levels (PM2.5 vs PM10)')

plt.show()
```

- X-axis → Pollutants
- Y-axis → Average value

- The bar chart compares average PM2.5 and PM10 levels.
- Each bar represents one pollutant.
- Taller bar indicates higher pollution level.
- Helps understand overall air quality condition.

## Global Air Pollution Trend Over Years (PM2.5 & PM10)

- Shows how global air pollution levels (PM2.5 and PM10) change over the years.

```python
yearly_pollution = df.groupby('year')[['pm2.5', 'pm10']].mean()

# Plot
plt.plot(yearly_pollution.index, yearly_pollution['pm2.5'], marker='o', label='PM2.5')
plt.plot(yearly_pollution.index, yearly_pollution['pm10'], marker='o', label='PM10')

plt.xlabel('Year')
plt.ylabel('Average Pollution Level')
plt.title('Global Air Pollution Trend Over Years')

plt.legend()
plt.show()
```

- X-axis → Year
- Y-axis → Average pollution level

- The graph shows the trend of air pollution over time using yearly averages.
- Two lines represent PM2.5 and PM10 pollution levels.
- An upward trend indicates increasing pollution levels.
- It helps compare how both pollutants change over time globally.

# Global Weather Condition Distribution

- Shows the percentage distribution of different weather conditions and identifies the most frequent one.

```python
condition_counts = df['condition_text'].value_counts()

top5 = condition_counts.head(5)
others = condition_counts[5:].sum()
top5['Others'] = others

plt.pie(top5, labels=top5.index, autopct='%1.1f%%')
plt.title('Weather Condition Distribution')
plt.show()
```

- Shows proportion of different weather conditions.

# Top 10 Countries with Highest Temperature

```python
country_temp=df.groupby('country')['celsius'].max()
top10=country_temp.sort_values(ascending=False).head(10)
plt.bar(top10.index,top10.values)
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('temperature(Cesius)')
plt.title('Top 10 Countries with Highest Temperature')
plt.show()
```

- X-axis : Countries
- Y-axis : Maximum Temperature(Celsius)

- The graph shows the top 10 countries with highest recoorded temperatures
- Taller bars represent countries with higher temperature values
- It helps compare extreme temperature conditions across countries

# Year-wise Humidity Trend of Top 10 Countries

```python
top_countries = df.groupby('country')['humidity'].mean().sort_values(ascending=False).head(10).index
data=df[df['country'].isin(top_countries)]
grouped=data.groupby(['year','country'])['humidity'].mean().unstack()
grouped.plot(marker='o')
plt.xlabel('Year')
plt.ylabel('Avarage Humidity(%)')
plt.title('Year-wise Humidity Trend of Top 10 Countries')
plt.legend(title='country',bbox_to_anchor=(1.05,1))
plt.show()
```

- X-axis : Year
- Y-axis : Avarage Humidity(%)

- The graph shows how humidity changes over years for the top 10 humid countries
- Each line represents one country
- It helps compare humidity trends across countries over time

## Temperature Distribution Across Locations 

- Shows how temperature varies across different locations and compares their distributions.

```python
top_locations = df['location'].value_counts().head(5).index

# Filter dataset
data = df[df['location'].isin(top_locations)]

# Prepare data for boxplot
box_data = [data[data['location'] == loc]['celsius'] for loc in top_locations]

# Plot
plt.boxplot(box_data, labels=top_locations)

plt.xlabel('Location')
plt.ylabel('Temperature (Celsius)')
plt.title('Temperature Distribution Across Locations')

plt.xticks(rotation=45)
plt.show()
```

- X-axis → Locations
- Y-axis → Temperature (Celsius)

- Each box represents temperature distribution for a location.
- The middle line shows the median temperature of that place.
- The box shows the range where most values lie.
- Points outside the box indicate extreme temperature values (outliers).

# Correlation Heatmap

## Correlation Heatmap of All Numerical Variables

- Shows the strength and direction of relationships between all numerical variables in the dataset.

```python
num_data = df.select_dtypes(include=['int64', 'float64'])

# Create heatmap
plt.figure(figsize=(25,15))
sns.heatmap(num_data.corr(), annot=True)  # use annot=True if you want values

plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

- The heatmap shows correlation between all numeric variables.
- Darker colors indicate stronger relationships.
- Positive correlation means variables increase together.
- Negative correlation means one increases while the other decreases.

## Correlation between Temperature and Feels Like Temperature (Heatmap)

- Shows how strongly temperature and feels like temperature are related.

```python
data = df[['celsius', 'feels_like_celsius']]

# Heatmap
plt.figure(figsize=(5,4))
sns.heatmap(data.corr(), annot=True)

plt.title('Temperature vs Feels Like Temperature Correlation')
plt.show()
```

- X-axis → Variables (Temperature, Feels Like Temperature)
- Y-axis → Variables

- The heatmap shows correlation values between the two variables.
- Values close to +1 indicate strong positive correlation.
- Since both are closely related, the value will be very high.
- It confirms that feels like temperature depends on actual temperature.

## Environmental Factors Correlation

- Correlation between Environmental Factors


```python
selected = df[['celsius','uv_index','visibility_km','pm2.5','humidity','precip_mm']]
corr=selected.corr()
plt.figsize=(6,5)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap of Environmental Factors')
plt.show()
```

- This heatmap shows the correlation between selected environmental factors
- Values close to +1 indicate positive correlation,and _1 indicate negative correlation
- It helps identify how weather conditions and pollution are related
